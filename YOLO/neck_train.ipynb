```python
%cd /
from glob import glob

img_list = glob('/content/drive/MyDrive/dmdkr_guitar/neck_dataset/images/*.jpg')
img_list = img_list+glob('/content/drive/MyDrive/dmdkr_guitar/neck_dataset/images/*.jpeg')

print(len(img_list))
```

```python
from sklearn.model_selection import train_test_split

train_img_list, val_img_list = train_test_split(img_list, test_size=0.25, random_state=42)

print(len(train_img_list), len(val_img_list))
```

```python
with open('/content/drive/MyDrive/dmdkr_guitar/neck_dataset/train.txt' , 'w') as f:
  f.write('\n'.join(train_img_list) + '\n')

with open('/content/drive/MyDrive/dmdkr_guitar/neck_dataset/val.txt', 'w') as f:
  f.write('\n'.join(val_img_list) + '\n')
```
