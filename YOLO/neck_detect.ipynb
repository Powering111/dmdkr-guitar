```python
#yolo 모델 적용

from IPython.display import Image
import shutil
import cv2
from PIL import Image
import numpy as np
import os
import glob

img_list = glob.iglob('/content/drive/MyDrive/dmdkr_guitar/code_video/frame/*/*.png',recursive=True)

i=1
flag = 0
for img in img_list:
  print(img)
  img_path = img

  save = img_path.split('frame/')
  save = save[1].split('/')[0]  

  !python /content/yolov5/detect.py --weights /content/drive/MyDrive/dmdkr_guitar/neck_dataset/YOLO_weight/best.pt --img 640 --conf 0.7 --source "{img}" --save-txt --save-crop
  
  yolo_det = glob.glob('/content/yolov5/runs/detect/exp'+str(i)+'/crops/neck/*.jpg')
  if i==1:
    yolo_det = glob.glob('/content/yolov5/runs/detect/exp/crops/neck/*.jpg')

  i=i+1
  print(yolo_det)
  if yolo_det != []:
    print(yolo_det)
    yolo_det = yolo_det[0]
    img = cv2.imread(yolo_det,cv2.IMREAD_GRAYSCALE)
    print(img.shape)

    h,w = img.shape
    if(h>w):
      percent = 500/h
    else:
      percent = 500/w
    
    resize_img = cv2.resize(img,dsize=(0,0),fx=percent,fy=percent)
    print(resize_img.shape)


    h,w = resize_img.shape
    margin = [np.abs(h - w) // 2, np.abs(h - w) // 2]
    if np.abs(h - w) % 2 != 0:
      margin[0] += 1
    if h < w:
      margin_list = [margin, [0, 0]]
    else:
      margin_list = [[0, 0], margin]
    
    pd_img = np.pad(resize_img, margin_list, mode='constant',constant_values=0)

    save = img_path.split('frame/')
    save = save[1].split('/')
    mkdir = '/content/drive/MyDrive/dmdkr_guitar/code_dataset3/'+save[0]
    save = '/content/drive/MyDrive/dmdkr_guitar/code_dataset3/'+save[0]+'/'+str(i-1)+'.jpg'
    print(save)

    if(not os.path.isdir(mkdir)):
      os.makedirs(mkdir)

    cv2.imwrite(save,pd_img)
```
